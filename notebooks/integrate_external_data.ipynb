```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

inventory_df = pd.read_csv("../data/Inventory_data.csv")

def scrape_data(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url)
        # Parse HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
                
        title = soup.find('title').get_text() if soup.find('title') else 'No title found'
        
        # Get description from meta tag
        description = ""
        meta_description = soup.find('meta', attrs={'name': 'description'})
        if meta_description:
            description = meta_description['content'].strip()
        
        
        return title, description
    
    except Exception as e:
        return None


# Let's take a small sample for demonstration purposes
sample_sites = inventory_df['site_name'].head(5)  # Use a small sample for demonstration

# Scrape information for each site
scraped_info = [scrape_data(f"https://{url}") or scrape_data(f"https://play.google.com/store/apps/details?id={url}") for url in sample_sites]



# Create a DataFrame from the scraped information
scraped_df = pd.DataFrame(scraped_info, columns=['site_title', 'site_description'])

# Assuming the sample_sites matches index-wise with scraped_df
enhanced_inventory_data = pd.concat([inventory_df.head(5).reset_index(drop=True), scraped_df], axis=1)
```
